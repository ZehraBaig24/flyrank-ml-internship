## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

 **Task Type:** Scoring

I chose **Refresh / Content Opportunity Scoring** as my ML task. The goal is to assign each webpage a score that represents how much it should be prioritized for content updates. This is a scoring problem because the model estimates how valuable it would be to refresh each page instead of making a simple yes/no prediction.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

 The target is a **content opportunity score** for each webpage. Since there is no direct label showing which pages should be refreshed, I would use a proxy based on observed search performance, such as impressions, clicks, CTR, and average position. The score is derived from these measured values rather than from a manually created rule.



## 3. Success metric

*One metric you can defend. What number means 'good'?*

I would use **Mean Absolute Error (MAE)** as the success metric because it measures how close the predicted scores are to the target scores. A lower MAE means the model predicts more accurate opportunity scores, helping content teams prioritize pages more effectively.
